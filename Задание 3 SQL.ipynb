```python
import pandas as pd
import numpy as np
import os
import warnings
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
data_folder = os.path.join(os.curdir,'./itunes_dataset/')
#створюємо датафрейм
df = pd.DataFrame()
for file in os.listdir(data_folder):
    if os.path.isfile(os.path.join(data_folder,file)) and os.path.splitext(os.path.join(data_folder,file))[1].lower() in ('.txt'):
        df = pd.concat([df, pd.read_csv(os.path.join(data_folder,file), sep="\t", index_col=None)])
```

```python
pd.set_option('display.max_columns', None)
df = df[df["Customer Currency"]=="USD"]
df_sub = df.drop_duplicates(subset=['Subscription Apple ID'])[['Subscription Apple ID','Subscription Name','Subscription Group ID','Subscription Duration','Introductory Price Type','Introductory Price Duration','Marketing Opt-In Duration','App Name','App Apple ID','Customer Price','Developer Proceeds']]
```

```python
df_sub = df_sub.where(pd.notnull(df_sub), None)
```

```python
df_sub['Subscription Apple ID'] = df_sub['Subscription Apple ID'].apply(str)
df_sub['Subscription Group ID'] = df_sub['Subscription Group ID'].apply(str)
df_sub['App Apple ID'] = df_sub['App Apple ID'].apply(str)
df_sub['Customer Price'] = df_sub['Customer Price'].apply(str)
df_sub['Developer Proceeds'] = df_sub['Developer Proceeds'].apply(str)
df_sub
```

# Запишем основные данные про подписки в Subscriptions ( можно руками 15 штук ) но сделаем это автоматизировано

```python
import pymysql
import pymysql.cursors
from os import getenv

HOST_MYSQL = '127.0.0.1'
PORT_MYSQL = 3306
USER_MYSQL = 'root'
PASSWORD_MYSQL = 'root'
DB_MYSQL = 'BetterMe'
CHARSET_MYSQL = 'utf8mb4'
CURSORCLASS_MYSQL = pymysql.cursors.DictCursor
```

```python
mysql_con = {
        "host": getenv("HOST_MYSQL", HOST_MYSQL),
        "port": getenv("PORT_MYSQL", PORT_MYSQL),
        "user": getenv("USER_MYSQL", USER_MYSQL),
        "password": (getenv("PASSWORD_MYSQL", PASSWORD_MYSQL)),
        "db": getenv("DB_MYSQL", DB_MYSQL),
        "charset": getenv("CHARSET_MYSQL", CHARSET_MYSQL),
        "cursorclass": getenv("CURSORCLASS_MYSQL", CURSORCLASS_MYSQL),
        }
```

```python
connection= pymysql.connect(**mysql_con)
```

```python
cursor=connection.cursor()
sql = "INSERT INTO Subscriptions(id, name, group_id, duration, type, price_duration, opt_in_duration,app_name,app_id,price,developer_proceeds) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for i,row in df_sub.iterrows():
    ### можно обернуть в try chatch что бы он игнорировал и не записывал данные если есть уже подписка
    cursor.execute(sql,tuple(row))
    connection.commit()
cursor.close()
```
